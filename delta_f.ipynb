```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
simulation = h5py.File('../bbh_upper_1000')
```

```python
mass1 = simulation['mass1'][:]
mass2 = simulation['mass2'][:]
tot_mass = mass1 + mass2
```

```python
dist = simulation['distance'][:]
```

### delta_f = 2.5

```python
data_2_5 = h5py.File('bbh_1000_2_5.h5', 'r')
data_2_5.keys()
```

```python
C40 = data_2_5['H1'][:]
C20 = data_2_5['L1'][:]
net_snr_2_5 = np.sqrt(C40**2 + C20**2)
```

```python
num_of_det = np.count_nonzero(net_snr_2_5 > 10)
num_of_det
```

```python
plt.scatter(tot_mass, net_snr_2_5, c=dist)
#plt.xscale('log')
plt.colorbar()
plt.show()
```
